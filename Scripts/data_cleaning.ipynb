```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("..\data\investments_VC.csv")
```

## Drop Values

```python
drop_cols = [
    'permalink',
    'name',
    'homepage_url'
]

df.drop(columns=drop_cols, inplace=True)
```

## Missing Values in Columns

```python
missing = (
    df.isnull()
      .sum()
      .to_frame("Missing Values")
)

missing["Percentage"] = (
    missing["Missing Values"] / len(df)
) * 100

missing.sort_values("Percentage", ascending=False)
```

## Handle Target Column

```python
unknown_df = df[df['status'].isna()].copy()

df = df[df['status'].isin(['acquired', 'closed'])].copy()
```

```python
# Remove leading and trailing whitespaces from all column names
df.columns = df.columns.str.strip()
```

```python
df['funding_total_usd'] = (
    df['funding_total_usd']
      .replace('-', np.nan)
      .str.replace(',', '', regex=False)
)

df['funding_total_usd'] = pd.to_numeric(
    df['funding_total_usd'],
    errors='coerce'
)
```

## Convert Date Column


```python
date_cols = [
    'founded_at',
    'first_funding_at',
    'last_funding_at'
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

## Verify Categorical Data

```python
for col in [
    'market',
    'country_code',
    'state_code'
]:
    print(df[col].nunique())
```

```python
# Converts 2026.0 to "2026" and leaves NaN empty
df['founded_year'] = df['founded_year'].astype('Int64').astype(str).str.replace('<NA>', '')
```

##### During data validation, 447 companies were identified where the first recorded funding date preceded the recorded founding date. Most discrepancies were within one year and are plausibly explained by pre-incorporation funding or differences between operational founding and legal incorporation. However, 12 companies exhibited gaps exceeding five years. These records were treated as anomalous and excluded from further analysis due to likely data quality issues.

### Remove above rows

```python
invalid_temporal = (
    (df["funding_rounds"] == 1) &
    (df["first_funding_at"] == df["last_funding_at"]) &
    ((df["founded_at"] - df["first_funding_at"]).dt.days > 365 * 5)
)

anomalies_df = df.loc[invalid_temporal].copy()
```

```python
df = df.drop(df.loc[invalid_temporal].index).reset_index(drop=True)
```
